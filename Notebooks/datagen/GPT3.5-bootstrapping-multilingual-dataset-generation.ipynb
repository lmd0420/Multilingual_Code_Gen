# get question


````python
import openai
import time
import pandas as pd
import re

openai.api_key = ""


def get_gpt_response(messages):
    try:
        completion = openai.ChatCompletion.create(
            model="gpt-3.5-turbo-0613", temperature=0.9, messages=messages
        )
        response_json_string = json.dumps(completion)
        response_dict = json.loads(response_json_string)
        content = response_dict["choices"][0]["message"]["content"].strip()
        return content

    except Exception as e:
        print(f"Error: {e}")
        print("Retrying in 30 seconds...")
        time.sleep(30)
        return get_gpt_response(messages)  # Recursive call to retry


def filter_questions(text):
    # Using regular expression to remove the initial numbers and extract questions
    questions = [
        re.sub(r"^\d+\.\s*", "", line).strip()
        for line in text.strip().split("\n")
        if line.strip()
    ]

    # Keeping only those questions that start with "Write a Python program"
    return [question for question in questions if question.startswith("Write")]


# Using a set to store unique questions
unique_questions = set()
cnt = 0
# Collecting English questions
while (
    len(unique_questions) < 300
):  # Adjust the condition to collect the desired number of unique questions
    print(cnt)
    messages = [
        {
            "role": "system",
            "content": "Generate 100 python problems (don't give me solution)",
        },
        {"role": "user", "content": "Generate a Python question."},
    ]
    response = get_gpt_response(messages)
    print(response)
    # Filter and add unique questions to the set
    questions = filter_questions(response)
    for question in questions:
        if question not in unique_questions:
            cnt += 1
            print(f"New question added: {question}")
            unique_questions.add(question)
        else:
            print(f"Duplicate question: {question}")

# Convert to DataFrame
df = pd.DataFrame(list(unique_questions), columns=["english_questions"])

# Print the final set of unique questions
print(df.head())
````

````python
# Save DataFrame to CSV
df.to_csv("unique_python_questions.csv", index=False)
````

# get python code answer


````python
def get_code_solution(question):
    try:
        messages = [
            {"role": "system", "content": "Provide answers in Python."},
            {"role": "user", "content": question},
        ]
        completion = openai.ChatCompletion.create(
            model="gpt-3.5-turbo-0613", temperature=0, messages=messages
        )
        response_json_string = json.dumps(completion)
        response_dict = json.loads(response_json_string)
        content = response_dict["choices"][0]["message"]["content"].strip()
        return content

    except Exception as e:
        print(f"Error: {e}")
        print("Retrying in 30 seconds...")
        time.sleep(30)
        return get_code_solution(question)  # Recursive call to retry


# Generate code solutions for the questions
questions_and_answers = []
cnt = 0
for question in unique_questions:
    print(cnt)
    code_answer = get_code_solution(question)
    print(question)
    print(code_answer)
    cnt += 1
    questions_and_answers.append((question, code_answer.strip()))

# Create a new DataFrame with questions and corresponding code answers
qa_df = pd.DataFrame(
    questions_and_answers, columns=["english_questions", "code_answers"]
)

# Print the questions and answers DataFrame
print(qa_df.head())

# Save the questions and answers DataFrame to a CSV file
qa_df.to_csv("qa_pairs_unfiltered.csv", index=False)
````

# extract python function


````python
import pandas as pd
import re

# Define the patterns for the extract_and_clean_code function
pattern = re.compile(r"```python(.*?)```", re.DOTALL)
assert_pattern = re.compile(r"^assert.*$", re.MULTILINE)
function_pattern = re.compile(r"(def .*?)(?:\n\n|$)", re.DOTALL)


# Function to extract and clean Python code
def extract_and_clean_code(text):
    match = pattern.search(text)
    if match:
        code = match.group(1).strip()
        cleaned_code = assert_pattern.sub(
            "", code
        ).strip()  # Remove lines starting with 'assert'

        # Extract only the function definition
        function_match = function_pattern.search(cleaned_code)
        if function_match:
            return function_match.group(1).strip()

    return "NO FUNCTION FOUND"


# Load the unfiltered question-code pairs from the CSV
qa_df = pd.read_csv("qa_pairs_unfiltered.csv")

# Apply the extract_and_clean_code function to clean up the code answers
qa_df["cleaned_code_answers"] = qa_df["code_answers"].apply(extract_and_clean_code)

# Optionally, filter out rows based on conditions (e.g., remove rows where no function definition was found)
qa_df = qa_df[qa_df["cleaned_code_answers"] != "NO FUNCTION FOUND"]


# Function to check if the code contains a complete Python function
def is_complete_python_function(code):
    return "def " in code and "return" in code


# Filter out rows that do not contain a complete Python function
qa_df = qa_df[qa_df["cleaned_code_answers"].apply(is_complete_python_function)]

# Print the filtered DataFrame
print(qa_df.head())

# Save the filtered question-code pairs to a new CSV file
qa_df.to_csv("qa_pairs_filtered.csv", index=False)
````

````python
# Keep only the 'english_questions' and 'cleaned_code_answers' columns
qa_df = qa_df[["english_questions", "cleaned_code_answers"]]

# Rename the columns
qa_df.columns = ["english_questions_filtered", "code_answers_filtered"]

# Print the filtered and renamed DataFrame
print(qa_df.head())

# Save the final DataFrame to a new CSV file
qa_df.to_csv("qa_pairs_final.csv", index=False)
````

# translate to others and translate back to eng

````python
# Function to perform translation and back-translation
def translate_and_back_translate(text, target_language):
    try:
        # Translate to the target language
        messages = [
            {"role": "system", "content": f"Translate the sentence into {target_language}."},
            {"role": "user", "content": text}
        ]
        completion = openai.ChatCompletion.create(
            model="gpt-3.5-turbo-0613",
            temperature=0,
            messages=messages
        )
        response_json_string = json.dumps(completion)
        response_dict = json.loads(response_json_string)
        translated_text = response_dict['choices'][0]['message']['content'].strip()
        print(text)
        print(translated_text)
        # Back translate to English
        messages = [
            {"role": "system", "content": "Translate the sentence into English."},
            {"role": "user", "content": translated_text}
        ]
        completion = openai.ChatCompletion.create(
            model="gpt-3.5-turbo-0613",
            temperature=0,
            messages=messages
        )
        response_json_string = json.dumps(completion)
        response_dict = json.loads(response_json_string)
        back_translated_text = response_dict['choices'][0]['message']['content'].strip()
        print(back_translated_text)
        return translated_text, back_translated_text
    except Exception as e:
        print(f"Error: {e}")
        return None, None

# Mount Google Drive
drive.mount('/content/drive')

# Load the filtered question-code pairs
qa_df = pd.read_csv('qa_pairs_final.csv')

# Define target languages and their codes
languages = ["Chinese", "Hindi", "Spanish", "Japanese", "Russian"]
language_map = {
    "Chinese": "zh-cn",
    "Hindi": "hi",
    "Spanish": "es",
    "Japanese": "ja",
    "Russian": "ru"
}

# Translate and back translate English questions, and add them to the DataFrame
for language in languages:
    print(f"Translating and back translating to {language} ({language_map[language]})...")
    translations, back_translations = zip(*qa_df['english_questions_filtered'].apply(lambda x: translate_and_back_translate(x, language)))
    qa_df[f'{language}_translation'] = translations
    qa_df[f'{language}_back_translation'] = back_translations

# Print the updated DataFrame with translations and back translations
print(qa_df.head())

# Save the translations and back translations to a new CSV file
translations_file = 'translations_and_back_translations.csv'
qa_df.to_csv(translations_file, index=False)

# Copy the CSV file to Google Drive
!cp {translations_file} "/content/drive/MyDrive/EACL/MultilingualLLMBias/results"
````

````python
from nltk.translate.bleu_score import sentence_bleu
import pandas as pd

# Load the translations and back translations from the CSV file
qa_df = pd.read_csv('/content/drive/MyDrive/EACL/MultilingualLLMBias/GPT3.5-finetune-data/translations_and_back_translations.csv')
# Define the target languages
languages = ["Chinese", "Hindi", "Spanish", "Japanese", "Russian"]

def calculate_bleu_score(original_text, back_translated_text):
    if isinstance(original_text, str) and isinstance(back_translated_text, str):
        return sentence_bleu([original_text.split()], back_translated_text.split(), weights=(0.5, 0.5))
    else:
        #print(f"Error: Original Text: {original_text}, Back Translated Text: {back_translated_text}")
        return 0

# Check BLEU score and retain translations if score is above the threshold
threshold = 0.8
rows_list = []

# Add original English questions to rows_list
english_count = 0
retained_counts = {language: 0 for language in languages}

for _, row in qa_df.iterrows():
    english_row = {'questions': row['english_questions_filtered'], 'code_answers': row['code_answers_filtered']}
    rows_list.append(english_row)
    english_count += 1

    # Check translations and retain if BLEU score exceeds the threshold
    for language in languages:
        bleu_score = calculate_bleu_score(row['english_questions_filtered'], row[f'{language}_back_translation'])
        if bleu_score > threshold:
            new_row = {'questions': row[f'{language}_translation'], 'code_answers': row['code_answers_filtered']}
            rows_list.append(new_row)
            retained_counts[language] += 1

# Print the count for each language including English
print(f"Number of original English questions: {english_count}")
for language, count in retained_counts.items():
    print(f"Number of retained translations for {language}: {count}")

stacked_df = pd.DataFrame(rows_list)

# Print the total number of rows in the stacked DataFrame
print(f"Total number of rows in the stacked DataFrame: {stacked_df.shape[0]}")

# Save the stacked DataFrame to a new CSV file
stacked_file = 'stacked_combined.csv'
stacked_df.to_csv(stacked_file, index=False)

# Copy the CSV file to Google Drive
!cp {stacked_file} "/content/drive/MyDrive/EACL/MultilingualLLMBias/GPT3.5-finetune-data"
````
